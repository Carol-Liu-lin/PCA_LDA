```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.utils import shuffle
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

from sklearn.datasets import load_iris    #导入数据集iris

#载入数据集  
iris = load_iris()  

def normlization(data):

    scaler = MinMaxScaler(feature_range=(-1,1))
    normalized_data = scaler.fit_transform(data)

    return normalized_data


def val_result(pipeline, X, Y, cv):                 #五折交叉验证, cv为折数
    accuracy=cross_val_score(pipeline, X, Y, cv=cv, scoring='accuracy').mean()
    precision_weighted=cross_val_score(pipeline, X, Y, cv=cv, scoring='precision_weighted').mean()
    recall_weighted=cross_val_score(pipeline, X, Y, cv=cv, scoring='recall_weighted').mean()
    f1_weighted=cross_val_score(pipeline, X, Y, cv=cv, scoring='f1_weighted').mean()
    print('accuracy: %f, precision weighted: %f, recall weighted: %f, f1_weighted: %f' %(accuracy, precision_weighted,
            recall_weighted, f1_weighted))
    return None

def model(type):
    if type == 'LR':
        return LogisticRegression()
    elif type == 'RF':
        return RandomForestClassifier(max_depth=2, random_state=0)
    elif type == 'DT':
        return DecisionTreeClassifier()
    elif type == 'KNN':
        return KNeighborsClassifier(n_neighbors=7)
    elif type == 'BG':
        return BaggingClassifier(base_estimator=LogisticRegression(),n_estimators=10, random_state=0)
    else:
        print(type + ' model is not supported!')


def preprocess(data):

    data = normlization(data)
    
    # PCA 
    #主成分分析法，返回降维后的数据
    #参数n_components为主成分数目
    #data = PCA(n_components=1).fit_transform(data)

    return data


def preprocess2(data,target):

    data = normlization(data)    
    
    #线性判别分析法，返回降维后的数据
    #参数n_components为降维后的维数
    data = LDA(n_components=3).fit_transform(data,target)

    return data


if __name__ == "__main__":
    
    
    train_data,test_data,train_y, test_y =train_test_split(iris.data, iris.target)
    #print(train_data, test_data,train_y, test_y)
    
    algorithm = 'KNN'

#     train_x = preprocess(train_data)
#     test_x = preprocess(test_data)
#     train_x, train_y = shuffle(train_x, train_y)
    
    train_x = preprocess2(train_data,train_y)
    test_x = preprocess2(test_data,test_y)
    train_x, train_y = shuffle(train_x, train_y)
    
    
    classifer = model(algorithm)
    val_result(classifer, train_x, train_y,5)          #五折交叉验证

    classifer.fit(train_x, train_y)
    _predicted_y_p = classifer.predict_proba(test_x)

    
```
